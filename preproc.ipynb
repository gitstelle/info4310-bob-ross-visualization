```python
def comb_col(Df, List, Col):
    '''
    Df: a pandas dataframe
    List: a list of columns (str) that you want to remove, belong to a general category
    Col: a string, the existing column that is the general category
    '''
    for x in List:
        Df.loc[Df[x]==1,Col] = 1
    
    Df = Df.drop(List, axis=1, inplace=True)
```
